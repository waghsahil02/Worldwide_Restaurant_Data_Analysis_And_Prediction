## Looking at above Value count, we get to know that for Aggregate rating 0, we are getting 2148 count, i.e 2148 rows having 0 aggregate rating. Also by observing the distribution plots and Box plots I get to know that all other values are having low frequency and only value 0 is having very high frequency i.e 2148, so in box plot 0 value is shown as an outlier. Now we can  also see that in "Rating Text" column 0 is represented as "Not Rated". 

## As per my information there are a few strategies we can consider to handle this issue:

- Imputation: Imputing values with Measures of Central tendency (Mean, Meadian, Mode)

- Separate Analysis: Performing separate analyses for the subset of data with actual ratings (i.e., non-zero ratings) and for the subset with missing or "Not Rated" values. 

- Drop Missing Values

## For now I choose to Impute 0 values with the Median values as Median is robust to outliers and skewed distributions.

```python
median = df[df['Aggregate rating'] != 0]['Aggregate rating'].median()

df['Aggregate rating'] = df['Aggregate rating'].replace(0, median)
```

```python
# Verify the changes
print(df['Aggregate rating'].value_counts())
```

## Task 3: Geospatial Analysis

- Visualize restaurant locations using latitude and longitude information, providing a spatial perspective.
- Analyze the geographical distribution of restaurants across different cities and countries.

### Visualize Locations of Restaurants

```python
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

gdf = gpd.GeoDataFrame(df,geometry=gpd.points_from_xy(df.Longitude, df.Latitude))

# Create a base map of the world using Geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Create a map that fits the screen and plots the restaurant locations
# The "continent" column is used for coloring and a legend is displayed
gdf.plot(ax=world.plot("continent", legend = True, figsize=(14, 12)), marker='o', color='red', markersize=15)

# Show the map
plt.show()
```

```python
plt.figure(figsize=(10, 6))

correlation = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()

sns.heatmap(correlation, annot=True, fmt=".2f")

plt.title("Correlation Between Restaurant's location and Rating")

plt.show()
```
